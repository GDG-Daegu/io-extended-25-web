### **1. Gemini 기반 웹페이지 구현 및 Chrome DevTools로 개선 (1시간)**

**활동 목표:** Gemini에게 웹페이지의 전체 HTML, CSS, JavaScript 코드를 생성하도록 요청하고, 그 결과물을 Chrome DevTools에 통합된 AI 기능을 사용해 실시간으로 디자인을 수정하고 개선합니다.

**실습 단계 (45분):**

1.  **시나리오 제안 및 코드 생성 (Gemini 사용):**

      * [gemini.google.com](https://gemini.google.com)에 접속.
      * **프롬프트 예시:** "Google I/O Extended 행사를 소개하는 랜딩 페이지를 만들어줘. 멋진 히어로 섹션, 세션 소개 섹션(AI, Web), 연사 소개 섹션, 그리고 간단한 등록 폼이 필요해. 모든 HTML, CSS, JavaScript 코드를 각각의 파일로 나누어 작성해줘."
      * Gemini가 생성한 코드를 `index.html`, `style.css`, `script.js` 파일로 로컬에 저장.

2.  **기본 웹페이지 확인:**

      * `index.html` 파일을 Chrome 브라우저에서 열어 초기 결과물을 확인.

3.  **Gemini in Chrome DevTools (가칭)로 디자인 개선:**

      * Chrome에서 개발자 도구(F12)를 오픈. (I/O 2025에서 관련 기능 발표를 가정)
      * CSS 패널이나 Elements 패널에 통합된 Gemini 아이콘 클릭.
      * **개선 요청 (자연어 프롬프트):**
          * "헤더를 스크롤해도 상단에 고정시켜줘. (Make the header sticky)"
          * "전체적인 컬러 테마를 구글의 머티리얼 블루 색상 톤으로 변경해줘."
          * "세션 카드에 마우스를 올리면 그림자 효과가 생기도록 애니메이션을 추가해줘."
      * Gemini가 제안하는 CSS/JS 코드 변경 사항을 실시간으로 확인하고 적용.

**랩업 및 Q\&A (15분):**

  * AI를 활용한 웹 개발의 장점(빠른 프로토타이핑, 코드 작성 시간 단축)과 주의할 점(생성된 코드 검토의 필요성)에 대해 토론.

### **2. Stitch & Builder.io: Low-Code로 CRM UI 제작 (45분)**

**활동 목표:** 드래그앤드롭 방식의 시각적 빌더인 Builder.io를 사용해 간단한 비즈니스 CRM 대시보드 UI를 디자인하고, 이를 실제 React 또는 HTML 코드로 추출하는 경험을 합니다. (*'Stitch'는 Builder.io에서 컴포넌트를 조합하는 과정을 의미하는 것으로 해석하여 진행합니다.*)

**컨셉 설명 (5분):**

  * **Builder.io:** 디자이너와 개발자가 협업할 수 있는 '시각적 헤드리스 CMS'. UI를 레고 블록처럼 조립(Stitch)하고, 완성된 디자인을 다양한 프레임워크의 코드로 즉시 변환할 수 있습니다.

**실습 단계 (Builder.io 웹사이트에서 진행) (40분):**

1.  **프로젝트 생성:**

      * [Builder.io](https://www.builder.io/)에 로그인하여 새 프로젝트(Space) 생성.

2.  **UI 컴포넌트 시각적으로 조립(Stitch):**

      * 'Visual Editor'에서 기본 제공되는 UI 컴포넌트(Box, Text, Image, Columns 등)를 드래그앤드롭.
      * **CRM 대시보드 UI 제작:**
          * 상단 헤더 (로고, 검색창, 사용자 프로필 아이콘)
          * 좌측 사이드바 (메뉴: 대시보드, 고객 목록, 리포트)
          * 중앙 콘텐츠 영역 (고객 목록을 보여주는 데이터 테이블)

3.  **스타일링 및 반응형 디자인:**

      * 우측의 스타일 탭에서 각 컴포넌트의 색상, 폰트, 간격 등 CSS 속성을 직접 수정.
      * 화면 상단의 디바이스 아이콘(데스크톱, 태블릿, 모바일)을 클릭하며 반응형 레이아웃을 조정.

4.  **코드로 변환 및 내보내기:**

      * 원하는 컴포넌트나 전체 페이지를 선택하고 **`Get Code`** 버튼 클릭.
      * 원하는 코드 형식(React, Vue, Svelte, Qwik, Solid, HTML/CSS)을 선택하고 생성된 코드를 확인.
      * 이 코드를 복사하여 로컬 코드 에디터에 붙여넣고 구조를 분석.

### **3. 추가 활동: Firebase로 배포하기 (15분)**

**활동 목표:** 1번에서 만든 랜딩 페이지 또는 2번에서 추출한 HTML 코드를 Firebase 호스팅을 통해 몇 가지 명령어로 전 세계 어디서든 접속 가능한 실제 웹사이트로 배포합니다.

**실습 단계 (터미널/명령 프롬프트에서 진행):**

1.  **Firebase 프로젝트 연결:**

      * 배포할 웹페이지 프로젝트 폴더로 이동.
      * `firebase login` 명령어로 구글 계정에 로그인.
      * `firebase init hosting` 명령어로 Firebase 프로젝트를 초기화.
          * `Use an existing project` 선택.
          * 웹사이트 콘텐츠가 있는 폴더(public directory)를 `.` 또는 `public`으로 지정.
          * `Configure as a single-page app?` 질문에 `No`로 답변 (정적 페이지이므로).

2.  **배포:**

      * `firebase deploy` 명령어 실행.
      * 배포가 완료되면 터미널에 **Hosting URL**이 표시됩니다.

3.  **결과 확인:**

      * 제공된 Hosting URL을 클릭하여 방금 배포한 나의 웹사이트에 접속\! 🥳
      * 참가자들과 함께 서로의 결과물을 공유하고 축하하는 시간을 갖습니다.
