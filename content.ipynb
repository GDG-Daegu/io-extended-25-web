#### **Step 3. 실시간으로 웹페이지 확인하고 개선하기**

0. ngrok 토큰을 설정해야 합니다. https://dashboard.ngrok.com/get-started/your-authtoken 에서 토큰을 복사해오세요.
1.  아래 코드를 실행하여 Colab 내부에 임시 웹 서버를 켜고, 외부에서 접속할 수 있는 \*\*공개 주소(Public URL)\*\*를 받으세요.
    ```python
    ngrok.set_auth_token("YOUR_NGROK_AUTHTOKEN") # ⬅️ 여기에 토큰을 붙여넣으세요.

    # public 폴더를 기준으로 웹 서버를 실행하고 공개 주소를 생성합니다.
    os.chdir('public')
    server_thread = threading.Thread(target=http.server.SimpleHTTPRequestHandler.server_forever, args=(("", 8000),))
    server_thread.daemon = True
    server_thread.start()
    public_url = ngrok.connect(8000)
    print(f"🚀 웹페이지 확인 URL: {public_url}")
    os.chdir('..') # 다시 상위 디렉토리로 이동
    ```
2.  출력된 `Public URL` 링크를 클릭하여 새 탭에서 방금 만든 웹사이트를 확인하세요.
