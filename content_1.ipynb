#### **Step 3. 실시간으로 웹 앱 확인하고 개선하기**

1.  **ngrok 토큰 설정 및 서버 실행:**
    * [ngrok 대시보드](https://dashboard.ngrok.com/get-started/your-authtoken)에 접속하세요.
    * **`Continue with Google`** 버튼을 클릭하여 구글 계정으로 로그인하세요.
    * 로그인 후 보이는 `Your Authtoken` 아래의 긴 문자열을 복사하여 아래 코드에 붙여넣습니다.
    * 아래 코드를 실행하여 웹 서버를 켜고 **공개 주소(Public URL)**를 받으세요.

```python
import os
import http.server
import threading
import time
from pyngrok import ngrok
import socket

# 1. 서버를 실행하는 함수 정의
def run_server(port=8000):
    """지정된 포트에서 HTTP 서버를 실행합니다."""
    server_address = ('', port)
    try:
        httpd = http.server.HTTPServer(server_address, http.server.SimpleHTTPRequestHandler)
        print(f"{port} 포트에서 서버를 시작합니다...")
        httpd.serve_forever()
    except OSError as e:
        if e.errno == 98:
            print(f"경고: {port} 포트가 이미 사용 중입니다. 다른 포트를 시도합니다...")
            run_server(port + 1)
        else:
            raise

# --- 메인 코드 ---
try:
    # ngrok 토큰 설정 시도 전 기존 터널 모두 제거
    ngrok.kill()
    time.sleep(1) # 프로세스가 완전히 종료될 때까지 잠시 대기

    # ngrok 인증 토큰 설정
    try:
        ngrok.set_auth_token("YOUT_NGROK_TOKEN")
        print("ngrok 토큰이 성공적으로 설정되었습니다.")
    except Exception as e:
        print(f"ngrok 토큰 설정 오류: {e}")
        # 토큰 설정에 실패하더라도 서버를 계속 실행하도록 합니다.
        pass

    # 웹 콘텐츠가 있는 public 폴더로 이동 (폴더가 없으면 생성)
    if not os.path.exists('public'):
        os.makedirs('public')
        print("public 폴더가 존재하지 않아 새로 생성했습니다.")

    # 현재 작업 디렉토리가 public 폴더인지 확인하고 이동
    if os.path.exists('public'):
        os.chdir('public')
    else:
        print("오류: 'public' 폴더를 찾을 수 없습니다. 스크립트를 중지합니다.")
        exit()

    # 2. 별도의 스레드에서 서버 실행 함수 호출
    server_thread = threading.Thread(target=run_server)
    server_thread.daemon = True
    server_thread.start()

    # 3. ngrok으로 공개 주소 생성
    try:
        public_url = ngrok.connect(8000)
        print(f"🚀 웹페이지 확인 URL: {public_url}")
    except Exception as e:
        print(f"ngrok 연결 오류: {e}")
        print("서버는 로컬에서 계속 실행됩니다. 터널링 없이 로컬에서만 접근 가능합니다.")
        public_url = "서버 URL을 가져올 수 없습니다."

    # 작업이 끝나면 원래 디렉토리로 돌아갑니다.
    os.chdir('..')

except Exception as e:
    print(f"스크립트 실행 중 예기치 않은 오류 발생: {e}")
```
